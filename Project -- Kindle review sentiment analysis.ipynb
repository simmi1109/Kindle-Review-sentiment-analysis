```python
import re
import nltk
from nltk.corpus import stopwords
from bs4 import BeautifulSoup
nltk.download('stopwords')
```

```python
## Removing special characters
df1['reviewText']=df1['reviewText'].apply(lambda x:re.sub('[^a-z A-z 0-9-]+', '',x))
## Remove the stopswords
df1['reviewText']=df1['reviewText'].apply(lambda x:" ".join([y for y in x.split() if y not in stopwords.words('english')]))
## Remove url 
df1['reviewText']=df1['reviewText'].apply(lambda x: re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '' , str(x)))
## Remove html tags
df1['reviewText']=df1['reviewText'].apply(lambda x: BeautifulSoup(x, 'lxml').get_text())
## Remove any additional spaces
df1['reviewText']=df1['reviewText'].apply(lambda x: " ".join(x.split()))
```

```python
df1.head()
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
df['reviewText']=df['reviewText'].apply(lambda x:lemmatize_words(x))
```

```python
### Train test split
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(df['reviewText'],df['rating'],test_size=0.2, random_state = 42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow =CountVectorizer()
X_train_bow =bow.fit_transform(X_train).toarray()
X_test_bow = bow.transform(X_test).toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
X_train_tf =bow.fit_transform(X_train).toarray()
X_test_tf = bow.transform(X_test).toarray()
```

### Model training

```python
from sklearn.naive_bayes import GaussianNB
nb_model_bow = GaussianNB().fit(X_train_bow,y_train)
nb_model_tfidf = GaussianNB().fit(X_train_tf,y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```
